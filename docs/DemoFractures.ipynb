# Example Usage Applied to Fracture Data - Åland Islands Networks

*Data from [Multi-scale fracture networks at Åland Islands - data analysis](https://github.com/nialov/multi-scale-fracture-networks-aland-islands-2022/tree/v0.3.0) by Nikolas Ovaskainen*

Fracture data and networks is another type of geospatial product which can use the Hough Transform method. Here we see mapped segments from bedrock in Åland Islands, Finland from [Ovaskainen et al., 2023](https://se.copernicus.org/articles/14/603/2023/#section3).

[![DOI](https://zenodo.org/badge/DOI/10.5281/zenodo.7919843.svg)](https://doi.org/10.5281/zenodo.7919843)


1. Load in a geoJSON format which also performs preprocessing.
2. Perform the [Hough Transform](https://en.wikipedia.org/wiki/Hough_transform)
3. Visualize the Results



```python
#Let's begin by loading the module
import linkinglines as ll

# Load the example dataset 
dikeset=ll.readFile('../data/example_getaberget_20m_9_traces.geojson', preprocess=True)
```

The next step is to perform the Hough Transform

Besides the data the input in the hough transform code is the center (xc,yc)
which is assumed to be the mean of the X and Y coordinates. It can be set arbitrarily. You may wish to set this to investigate structures in your dataset. But most people will use the default settings.


```python

dikeset, xc, yc=ll.HoughTransform(dikeset)

print(dikeset.columns)
```

Visualize the hough transform data.

```python
fig,ax=ll.DotsLines(dikeset, ColorBy='theta')

fig2, ax2 = ll.plotScatterHist(dikeset, "theta", "rho")
```
